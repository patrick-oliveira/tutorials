```python
class SeriesDataset(Dataset):
    def __init__(self, dataframe):
        super(SeriesDataset, self)
        self.dataframe = dataframe
        self.inputs = self.dataframe.iloc[:, :-1]
        self.targets = self.dataframe.iloc[:, -1]
        
    def __len__(self):
        return len(self.dataframe)
    
    def __getitem__(self, idx):
        return (torch.tensor(self.inputs.loc[idx].values, dtype = torch.float).reshape(-1, 1), \
                torch.tensor(self.targets.loc[idx], dtype = torch.float))
```

```python
dataset = SeriesDataset(data)
batch_size = 300
dataloader = DataLoader(dataset, shuffle = True, batch_size = batch_size)
```

```python
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(RNN, self).__init__()
        self.hidden_size = hidden_size
        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first = True)
        self.output = nn.Linear(hidden_size*num_layers, 1)
        
    def forward(self, x, hidden):
        batch_size = x.shape[0]
        out, hidden = self.rnn(x, hidden)
        out = hidden.transpose(0, 1).reshape(batch_size, -1)
        out = self.output(out)
        out = out.view(-1)
        return out, hidden
    
def train_rnn(model, dataloader, loss_function, optimizer, epochs):
    mean = lambda xs: sum(xs)/len(xs)
    epoch_losses = []
    for i in range(1, epochs+1):
        losses = []
        for inputs, targets in dataloader: # how the batches are obtained?
                optimizer.zero_grad()
                hidden = None
                pred, hidden = model(inputs, hidden)
                
                loss = loss_function(pred, targets)
                losses.append(loss.item())
                
                loss.backward(retain_graph = True)
                optimizer.step()
                
        epoch_losses.append(mean(losses))
        if i%10 == 0:
            print('Epoch: {}/{}.............'.format(i, epochs), end=' ')
            print("Loss: {:.4f}".format(epoch_losses[-1]))
    
    return epoch_losses
```

```python
model = RNN(1, 10, 3)
criterion = nn.MSELoss()
lr = 0.01
optimizer = torch.optim.Adam(model.parameters(), lr = lr)
epochs = 200
```

```python
losses = train_rnn(model, dataloader, criterion, optimizer, epochs)
```

```python
x_range = range(len(losses))
plt.plot(x_range, losses)
plt.xlabel("epochs")
plt.ylabel("Loss function")
plt.show()
```
